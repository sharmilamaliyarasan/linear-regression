```python
import pandas as pd
from sklearn.linear_model import LinearRegression
df = pd.read_csv("regression_datasets_large.xlsx - Electricity_Bill.csv")
X = df[["Units_Consumed"]]
y = df["Bill_Amount"]
model = LinearRegression()
model.fit(X, y)
predicted_bill = model.predict(pd.DataFrame([[350]], columns=["Units_Consumed"]))[0]
fixed_charge = model.intercept_
print("Predicted Bill for 350 units:", predicted_bill)
print("Fixed Charge (Intercept):", fixed_charge)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
df = pd.read_csv("regression_datasets_large.xlsx - Movie_Revenue.csv")
X = df[["Marketing_Spend_Cr"]]
y = df["Box_Office_Revenue_Cr"]
model = LinearRegression()
model.fit(X, y)
predicted_revenue = model.predict(pd.DataFrame([[10]], columns=["Marketing_Spend_Cr"]))[0]
print("Predicted Revenue for ₹10 Cr spend:", predicted_revenue)
print("Fixed Component (Intercept):", model.intercept_)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
df = pd.read_csv("regression_datasets_large.xlsx - Cold_Drink_Sales.csv")
X = df[["Temperature_C"]]
y = df["Sales"]
model = LinearRegression()
model.fit(X, y)
predicted_sales = model.predict([[40]])[0]
print("Predicted Sales at 40°C:", predicted_sales)
residuals = y - model.predict(X)
plt.scatter(X, residuals)
plt.axhline(y=0, color="red")
plt.xlabel("Temperature (°C)")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df = pd.read_csv("regression_datasets_large.xlsx - Calories_Burned.csv")
X = df[["Duration_Minutes"]]
y = df["Calories_Burned"]

model = LinearRegression()
model.fit(X, y)

calories_1hr = model.predict(pd.DataFrame([[60]], columns=["Duration_Minutes"]))[0]
print("Estimated Calories Burned in 1 hour:", calories_1hr)

residuals = y - model.predict(X)
plt.scatter(X, residuals)
plt.axhline(y=0, color="red")
plt.xlabel("Duration (minutes)")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```
